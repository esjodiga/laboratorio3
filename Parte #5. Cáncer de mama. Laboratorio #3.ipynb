```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn.metrics import confusion_matrix
import matplotlib.patches as mpatches
import seaborn as sb
from sklearn.metrics import classification_report
from sklearn import preprocessing
```

```python
df_diagnostico = pd.read_csv("bdiag.csv",sep=",")

df_diagnostico
```

```python
df_diagnostico.describe()
```

```python
print(df_diagnostico.groupby("diagnosis").size())
```

```python
df_diagnostico.columns
```

```python
#Convertimos la variable diagnosis en una dummy, en donde "benign, B" = 0 y "malign, M" = 1:
df_diagnostico.loc[df_diagnostico["diagnosis"] == "B", "diagnosis"] = 0
df_diagnostico.loc[df_diagnostico["diagnosis"] == "M", "diagnosis"] = 1

df_diagnostico[["diagnosis"]] = df_diagnostico[["diagnosis"]].astype(int)
```

```python
df_diagnostico
```

```python
df_diagnostico.corr()
```

```python
x = df_diagnostico[["radius_mean","perimeter_mean","area_mean",
                    "concave points_mean","radius_worst","perimeter_worst",
                    "area_worst","concave points_worst"]]
#se eligieron las variables independientes de acuerdo a su nivel de correlación con diagnosis
y = df_diagnostico['diagnosis']

x = preprocessing.StandardScaler().fit(x).transform(x.astype(float))

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=4)

print(x_train.shape)
print(x_test.shape)

print(y_train.shape)
print(y_test.shape)
```

```python
k=5
diagnostico_cancer_pred = KNeighborsClassifier(n_neighbors = k).fit(x_train,y_train) #entrenamos el modelo
y_hat = diagnostico_cancer_pred.predict(x_test) #lo corremos con el test de prueba
```

```python
print(confusion_matrix(y_test, y_hat)) #Vemos que 107 de 114 datos fueron clasificados correctamente
print(classification_report(y_test, y_hat))
```

```python
k_range = range(1,100)
certeza = {}
certeza_list = []

for k in k_range:
    knn = KNeighborsClassifier(n_neighbors = k).fit(x_train,y_train)
    y_hat = knn.predict(x_test)
    certeza[k] = metrics.accuracy_score(y_test,y_hat)
    certeza_list.append(metrics.accuracy_score(y_test,y_hat))
    
plt.plot(k_range,certeza_list)
plt.xlabel("Valores K")
plt.ylabel("Certeza del modelo")
plt.show()

print(certeza_list) #vemos que el k=5 elegido no es el que tiene la mejor certeza pero sí un nivel adecuado, por encima del 93%
#Sin embargo, es un buen estimador
```
